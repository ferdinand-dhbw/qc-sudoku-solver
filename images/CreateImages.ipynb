```python
from PIL import Image, ImageDraw, ImageFont
import imageio
import os

SIZE = 3

def create_image(fixed, highlight, filename, c1="Yellow", c2="Orange", special=None, c3=None):
    
    directory = os.path.dirname(filename)
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    total_width = 318
    total_height = 318
    
    font = ImageFont.truetype("C:\Windows\Fonts\\ariblk.ttf", 16)
    
    img = Image.new("RGB", (total_width, total_height), color="white")

    d = ImageDraw.Draw(img)
    
    #d.rectangle([(4,4), (total_width - 4, total_height - 4)], outline="Red")

    width = 20
    height = 20
    gap=8

    width2 = int(width/2)
    height2 = int(height/2)
    gap2 = int(gap/2)
    
    min_x = 9999999
    min_y = 9999999
    
    max_x = 0
    max_y = 0

    v=0
    c=0
    for r in range(SIZE):
        x_base = 60 + c*(width+gap) + v*(width2+gap2)
        y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        d.ellipse([(x_base - 40, y_base - 30),(x_base - 10, y_base)], outline="Blue", fill="white")        
        (t_width, t_height) = d.textsize(str(r),font=font) 
        d.text(((x_base - 40) + (15 - t_width//2), (y_base - 30) + (15 - t_height//2)), str(r), fill="black", font=font) 
        

    r=0
    v=0
    for c in range(SIZE):
        x_base = 60 + c*(width+gap) + v*(width2+gap2)
        y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        d.ellipse([(x_base, y_base + 10),(x_base + 30, y_base + 40)], outline="Red", fill="white")
        (t_width, t_height) = d.textsize(str(c),font=font) 
        d.text(((x_base) + (15 - t_width//2),( y_base + 10) + (15 - t_height//2)), str(c), fill="black", font=font)
       
    r=0
    c=SIZE-1
    for v in range(SIZE*SIZE-1, -1, -1):
        x_base = 60 + c*(width+gap) + v*(width2+gap2)
        y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        d.ellipse([(x_base + 30, y_base + 0),(x_base + 30 + 30, y_base + 0 + 30)], outline="Green", fill="white")        
        (t_width, t_height) = d.textsize(str(v),font=font) 
        d.text(((x_base + 30) + (15 - t_width//2), (y_base + 0) + (15 - t_height//2)), str(v), fill="black", font=font)
        

    for v in range(SIZE*SIZE-1, -1, -1):
        for r in range(SIZE):
            for c in range(SIZE):
                
                x_offset = 0
                y_offset = 0
                
                x_base = 60 + x_offset + c*(width+gap) + v*(width2+gap2)
                y_base = 174 + y_offset + (SIZE-r)*(height+gap) - v*(height2+gap2)

                color1 = "Ivory"
                color2 = "Snow"

                for index in range(len(fixed)):
                    if r in fixed[index][0] and c in fixed[index][1] and v in fixed[index][2]:
                        color1 = c1
                        color2 = color1

                for index in range(len(highlight)):
                    if r == highlight[index][0] and c == highlight[index][1] and v == highlight[index][2]:
                        color1 = c2 
                        color2 = color1

                if special is not None:
                    if r == special[0] and c == special[1] and v == special[2]:
                        color1 = c3 
                        color2 = color1


                A = (x_base, y_base)
                B = (x_base + width, y_base)
                C = (x_base + width, y_base - height)
                D = (x_base, y_base-height)
                E = (x_base + width2 + width, y_base - height2)
                F = (x_base + width2 + width, y_base - height - height2)
                G = (x_base + width2, y_base - height - height2)

                d.rectangle([A, C], outline="Blue", fill=color1)
                d.polygon([D, G, F, C], outline="Blue", fill=color1)
                d.polygon([C, F, E, B], outline="Blue", fill=color2)


                TL = (x_base, y_base - height - height2)
                BR = (x_base + width2 + width, y_base)

                min_x = min(min_x, TL[0])
                max_x = max(max_x, BR[0])
                min_y = min(min_y, TL[1])
                max_y = max(max_y, BR[1])
                
    img.save(filename)
    
#-----------------------------------------------------------

filenames = []
filename = ".\\bits-overview.png" 
create_image([([], [], [])], [(-1, -1, -1)], filename)
filenames.append(imageio.imread(filename))
imageio.mimsave(".\\bits-overview.gif", filenames)

#-----------------------------------------------------------

filenames = []
for v in range(SIZE*SIZE):
    filename = ".\\cell-single\\%d.png" % v
    create_image([([SIZE-1], [SIZE-1], range(SIZE*SIZE))], [(SIZE-1, SIZE-1, v)], filename)
    filenames.append(imageio.imread(filename))
imageio.mimsave(".\\cell-single.gif", filenames, duration=0.1)

#-----------------------------------------------------------

filenames = []
for r in range(SIZE):
    for c in range(SIZE):
        filename = ".\\cell-all\\%d-%d.png" % (r, c)
        create_image([([r], [c], range(SIZE*SIZE))], [(-1, -1, -1)], filename)
        filenames.append(imageio.imread(filename))
imageio.mimsave(".\\cell-all.gif", filenames, duration=0.5)

#-----------------------------------------------------------

filenames = []
for c in range(SIZE):
    for v in range(SIZE*SIZE):
        filename = ".\\row-single\\%d-%d.png" % (c,v)
        create_image([([SIZE-1], range(SIZE), range(SIZE*SIZE))], [(SIZE-1, c, v)], filename)
        filenames.append(imageio.imread(filename))
imageio.mimsave(".\\row-single.gif", filenames, duration=0.1)

#-----------------------------------------------------------

filenames = []
for r in range(SIZE):
    filename = ".\\row-all\\%d.png" % r
    create_image([([r], range(SIZE), range(SIZE*SIZE))], [(-1, -1, -1)], filename)
    filenames.append(imageio.imread(filename))
imageio.mimsave(".\\row-all.gif", filenames, duration=0.5)

#-----------------------------------------------------------

filenames = []
for r in range(SIZE):
    for v in range(SIZE*SIZE):
        filename = ".\\column-single\\%d-%d.png" % (r,v)
        create_image([(range(SIZE),[SIZE - 1], range(SIZE*SIZE))], [(r, SIZE - 1, v)], filename)
        filenames.append(imageio.imread(filename))
imageio.mimsave(".\\column-single.gif", filenames, duration=0.1)

#-----------------------------------------------------------

filenames = []
for c in range(SIZE):
    filename = ".\\column-all\\%d.png" % c
    create_image([(range(SIZE),[c], range(SIZE*SIZE))], [(-1, -1, -1)], filename)
    filenames.append(imageio.imread(filename))
imageio.mimsave(".\\column-all.gif", filenames, duration=0.5)

#-----------------------------------------------------------

filenames = []
filename = ".\\diagonal_1.png" 
create_image([([i],[i], range(SIZE*SIZE)) for i in range(SIZE)], [(-1, -1, -1)], filename)
filenames.append(imageio.imread(filename))
imageio.mimsave(".\\diagonal_1.gif", filenames)

#-----------------------------------------------------------

filenames = []
filename = ".\\diagonal_2.png" 
create_image([([i],[SIZE-i-1], range(SIZE*SIZE)) for i in range(SIZE)], [(-1, -1, -1)], filename)
filenames.append(imageio.imread(filename))
imageio.mimsave(".\\diagonal_2.gif", filenames)

#-----------------------------------------------------------

print("ready")
```
