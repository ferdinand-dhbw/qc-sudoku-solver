```python
from PIL import Image, ImageDraw, ImageFont
import imageio
import os

SIZE = 9

def create_image(fixed, highlight, filename, c1="Yellow", c2="Orange", special=None, c3=None):
    
    total_width = 486
    total_height = 486
    
    font = ImageFont.truetype("C:\Windows\Fonts\\ariblk.ttf", 16)
    
    img = Image.new("RGB", (total_width, total_height), color="white")

    d = ImageDraw.Draw(img)
    
    #d.rectangle([(4,4), (total_width - 4, total_height - 4)], outline="Red")

    width = 20
    height = 20
    gap=8

    width2 = int(width/2)
    height2 = int(height/2)
    gap2 = int(gap/2)
    
    min_x = 9999999
    min_y = 9999999
    
    max_x = 0
    max_y = 0

    v=0
    c=0
    for r in range(SIZE):
        x_base = 60 + c*(width+gap) + v*(width2+gap2)
        y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        d.ellipse([(x_base - 40, y_base - 30),(x_base - 10, y_base)], outline="Blue", fill="white")        
        (t_width, t_height) = d.textsize(str(r),font=font) 
        d.text(((x_base - 40) + (15 - t_width//2), (y_base - 30) + (15 - t_height//2)), str(r+1), fill="black", font=font) 
        

    r=0
    v=0
    for c in range(SIZE):
        x_base = 60 + c*(width+gap) + v*(width2+gap2)
        y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        d.ellipse([(x_base, y_base + 10),(x_base + 30, y_base + 40)], outline="Red", fill="white")
        (t_width, t_height) = d.textsize(str(c),font=font) 
        d.text(((x_base) + (15 - t_width//2),( y_base + 10) + (15 - t_height//2)), str(c+1), fill="black", font=font)
       
    r=0
    c=SIZE-1
    for v in range(SIZE-1, -1, -1):
        x_base = 60 + c*(width+gap) + v*(width2+gap2)
        y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        d.ellipse([(x_base + 30, y_base + 0),(x_base + 30 + 30, y_base + 0 + 30)], outline="Green", fill="white")        
        (t_width, t_height) = d.textsize(str(v),font=font) 
        d.text(((x_base + 30) + (15 - t_width//2), (y_base + 0) + (15 - t_height//2)), str(v+1), fill="black", font=font)
        

    if True:    
        for v in range(SIZE-1, -1, -1):
            for r in range(SIZE):
                for c in range(SIZE):
        
                    x_base = 60 + c*(width+gap) + v*(width2+gap2)
                    y_base = 174 + (SIZE-r)*(height+gap) - v*(height2+gap2)
        
                    color1 = "Ivory"
                    color2 = "Snow"
            
                    for index in range(len(fixed)):
                        if r in fixed[index][0] and c in fixed[index][1] and v in fixed[index][2]:
                            color1 = c1
                            color2 = color1
                        
                    for index in range(len(highlight)):
                        if r == highlight[index][0] and c == highlight[index][1] and v == highlight[index][2]:
                            color1 = c2 
                            color2 = color1
                            
                    if special is not None:
                        if r == special[0] and c == special[1] and v == special[2]:
                            color1 = c3 
                            color2 = color1
                        
                    
                    A = (x_base, y_base)
                    B = (x_base + width, y_base)
                    C = (x_base + width, y_base - height)
                    D = (x_base, y_base-height)
                    E = (x_base + width2 + width, y_base - height2)
                    F = (x_base + width2 + width, y_base - height - height2)
                    G = (x_base + width2, y_base - height - height2)
        
                    d.rectangle([A, C], outline="Blue", fill=color1)
                    d.polygon([D, G, F, C], outline="Blue", fill=color1)
                    d.polygon([C, F, E, B], outline="Blue", fill=color2)

                
                    TL = (x_base, y_base - height - height2)
                    BR = (x_base + width2 + width, y_base)
                
                    min_x = min(min_x, TL[0])
                    max_x = max(max_x, BR[0])
                    min_y = min(min_y, TL[1])
                    max_y = max(max_y, BR[1])
                
    img.save(filename)
    
    #print("min_x = %d" % min_x)
    #print("max_x = %d" % max_x)
    #print("min_y = %d" % min_y)
    #print("max_y = %d" % max_y)
    #print("")

#-----------------------------------------------------------

for x1 in ["cell", "row", "column", "block", "hint"]:
    directory = ".\\%s" % x1
    if not os.path.exists(directory):
        try:
            os.makedirs(directory)
        except OSError:
            print("Creation of the directory %s failed" % directory)

    
    for x2 in ["A", "B", "C"]:
        directory = ".\\%s\\%s" %(x1, x2)
        if not os.path.exists(directory):
            try:
                os.makedirs(directory)
            except OSError:
                print("Creation of the directory %s failed" % directory)

        
#-----------------------------------------------------------

filename = ".\\bits-overview.png" 
create_image([([], [], [])], [(-1, -1, -1)], filename)

create_cell_images= True
create_row_images= True
create_column_images= True
create_block_images= True
create_hint_images= True
    
#-----------------------------------------------------------

if create_cell_images:
    images = []
    for v in range(SIZE):
        filename = ".\\cell\\A\\%d.png" % v
        create_image([([8], [8], range(SIZE))], [(8, 8, v)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\cell\\A.gif", images, duration=0.6)

    images = []
    counter=0
    rc_list = []
    for i in range(SIZE):
        for j in range(i):
            rc_list.append((SIZE-1-i, SIZE-1-j))  
        rc_list.append((SIZE-1-i, SIZE-1-i))
        for j in range(i):
            rc_list.append((SIZE-i+j, SIZE-1-i))            

    for (r,c) in rc_list:
        filename = ".\\cell\\B\\%d.png" % counter
        counter+=1
        create_image([([r], [c], range(SIZE))], [(-1, -1, -1)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\cell\\B.gif", images, duration=0.2)

#-----------------------------------------------------------

if create_row_images:
    images = []
    for c in range(SIZE):
        filename = ".\\row\\A\\%d.png" % c
        create_image([([0], range(SIZE), [0])], [(0, c, 0)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\row\\A.gif", images, duration=0.6)

    images = []
    for r in range(SIZE):
        filename = ".\\row\\B\\%d.png" % r
        create_image([([r], range(SIZE), [0])], [(-1, -1, -1)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\row\\B.gif", images, duration=0.6)

    images = []
    for v in range(SIZE):
        filename = ".\\row\\C\\%d.png" % v
        create_image([(range(SIZE), range(SIZE), [v])], [(-1, -1, -1)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\row\\C.gif", images, duration=0.6)


#-----------------------------------------------------------

if create_column_images:
    images = []
    for r in range(SIZE):
        filename = ".\\column\\A\\%d.png" % r
        create_image([(range(SIZE),[0], [0])], [(r, 0, 0)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\column\\A.gif", images, duration=0.6)

    images = []
    for c in range(SIZE):
        filename = ".\\column\\B\\%d.png" % c
        create_image([(range(SIZE),[c], [0])], [(-1, -1, -1)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\column\\B.gif", images, duration=0.6)

    images = []
    for v in range(SIZE):
        filename = ".\\column\\C\\%d.png" % v
        create_image([(range(SIZE), range(SIZE), [v])], [(-1, -1, -1)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\column\\C.gif", images, duration=0.6)

#-----------------------------------------------------------

if create_block_images:
    images = []
    for i in range(SIZE):
        filename = ".\\block\\A\\%d.png" % i
        create_image([([0,1,2], [0,1,2], [0])], [(i//3, i%3, 0)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\block\\A.gif", images, duration=0.6)

    images = []
    counter=0
    for f_r in [[0,1,2], [3,4,5], [6,7,8]]:
        for f_c in [[0,1,2], [3,4,5], [6,7,8]]:
            filename = ".\\block\\B\\%d.png" % counter
            create_image([(f_r, f_c, [0])], [(-1, -1, -1)], filename)
            counter +=1
            images.append(imageio.imread(filename))
    imageio.mimsave(".\\block\\B.gif", images, duration=0.6)

    images = []
    for v in range(SIZE):
        filename = ".\\block\\C\\%d.png" % v
        create_image([(range(SIZE),  range(SIZE), [v])], [(-1, -1, -1)], filename)
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\block\\C.gif", images, duration=0.6)

#-----------------------------------------------------------

if create_hint_images:
    images = []
    for v in range(1, SIZE):
        filename = ".\\hint\\A\\%2d.png" % len(images)
        create_image([
                        ([8],  range(SIZE), [0]),
                        (range(SIZE), [8], [0]),
                        ([8], [8], range(SIZE)),
                        ([7],  [6,7,8], [0]),
                        ([6],  [6,7,8], [0]),
                     ], 
                     [(8, 8, v)], filename, c1="silver", c2="grey", special=(8,8,0), c3="green")
        images.append(imageio.imread(filename))
    for c in range(1, SIZE):
        filename = ".\\hint\\A\\%2d.png" % len(images)
        create_image([
                        ([8],  range(SIZE), [0]),
                        (range(SIZE), [8], [0]),
                        ([8], [8], range(SIZE)),
                        ([7],  [6,7,8], [0]),
                        ([6],  [6,7,8], [0]),
                     ], 
                     [(8, SIZE - 1 - c, 0)], filename, c1="silver", c2="grey", special=(8,8,0), c3="green")
        images.append(imageio.imread(filename))
    for r in range(1, SIZE):
        filename = ".\\hint\\A\\%2d.png" % len(images)
        create_image([
                        ([8],  range(SIZE), [0]),
                        (range(SIZE), [8], [0]),
                        ([8], [8], range(SIZE)),
                        ([7],  [6,7,8], [0]),
                        ([6],  [6,7,8], [0]),
                     ], 
                     [(SIZE - 1 - r, 8, 0)], filename, c1="silver", c2="grey", special=(8,8,0), c3="green")
        images.append(imageio.imread(filename))
    for i in range(1, SIZE):
        filename = ".\\hint\\A\\%2d.png" % len(images)
        create_image([
                        ([8],  range(SIZE), [0]),
                        (range(SIZE), [8], [0]),
                        ([8], [8], range(SIZE)),
                        ([7],  [6,7,8], [0]),
                        ([6],  [6,7,8], [0]),
                     ], 
                     [(6 + (SIZE - 1 - i)//3, SIZE - 1 - i%3, 0)], filename, c1="silver", c2="grey", special=(8,8,0), c3="green")
        images.append(imageio.imread(filename))
    imageio.mimsave(".\\hint\\a.gif", images, duration=0.6)

#-----------------------------------------------------------

print("ready")
```
